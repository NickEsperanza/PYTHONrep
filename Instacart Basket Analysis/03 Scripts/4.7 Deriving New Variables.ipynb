```python
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
df_ords_prods_merged = df_ords_prods_merged[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df_ords_prods_merged['price_range'] = df_ords_prods_merged.apply(price_label, axis=1)
```

# Busiest Days now includes Saturday (0) and Sunday (1). Slowest Days now includes Tuesday (3) and Wednesday (4).

```python
result = []

for value in df_ords_prods_merged["orders_days_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1:
      result.append("Busiest day")
  elif value == 3:
    result.append("Slowest day")
  elif value == 4:
    result.append("Slowest day")
  else:
    result.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
df_ords_prods_merged['Busiest_Days'] = result
```

# Designated values to each category for Most Orders, Average Orders, and Fewest Orders

```python
result = []
for value in df_ords_prods_merged["hour_of_day_ordered"]:
    if value in (9,10,11,12,13,14,15,16,17):
        result.append("Most Orders")
    elif value in (8,18,19,20,7,21,22,23,6):
        result.append("Average Orders")
    elif value in (0,1,5,2,3,4):
        result.append("Fewest Orders")
```
