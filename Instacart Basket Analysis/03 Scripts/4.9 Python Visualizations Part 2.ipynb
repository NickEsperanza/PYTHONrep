# Creating a sample for price and hour of the day and order is made

```python
np.random.seed(4)
dev = np.random.rand(len(new_customer_data)) <= 0.7
```

```python
# store 70% of data in big
big = new_customer_data[dev]
```

```python
# store 30% of data in small
small = new_customer_data[~dev]
```

```python
# short cut to only include necessary data for line graph
customer_2 = small[['prices','hour_of_day_ordered']]
```

```python
# line graph reflecting prices and hour of the day ordered for items
line = sns.lineplot(data = customer_2, x = 'hour_of_day_ordered',y = 'prices')
```

```python
# creating a shortcut to view relationship between age and number of dependants
customer_3 = small[['Age','Dependants']]
```

```python
# creating a line graph reflecting age and number of dependants
line = sns.lineplot(data = customer_3, x = 'Age',y = 'Dependants')
```

# The above line graph shows that there is no true correlation between age and number of dependants. 

```python
# creating a scatterplot to determine connections between age and income
sns.scatterplot(x = 'Age', y = 'income',data = new_customer_data)
```

# The above graph shows there is some correlation between age and income. While a heavy majority across all ages fall within an income of $20,000, we see a separation start to occur after age 40. After age 40, higher income earners at $50,000 and more start to appear where younger age groups have no recorded earners in the same brackets.
